# Import Libraries

```python
# General
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

import os
import pickle
```

# Model Inference

Pada notebook ini model akan dilakukan pengujian dengan data inference atau unseen data untuk menguji performa model. Data ini sudah di ambil sebelumnya setelah proses EDA pada notebook utama.

## Load Data

```python
# Load data
df_inf = pd.read_csv('Inf_Data.csv')
df_inf
```

## Load Model

```python
# Tentukan path ke file model.pkl
model_path = os.path.join('deployment', 'model.pkl')

# Muat model dari file
with open(model_path, 'rb') as file_1:
    model = pickle.load(file_1)
```

```python
model
```

```python
# Predicting
y_pred_inf = model.predict(df_inf)

# Menampilkan prediksi
if y_pred_inf[0] == 1:
    print('Karyawan tidak meninggalkan perusahaan.')
else:
    print('Karyawan meninggalkan perusahaan.')
```
